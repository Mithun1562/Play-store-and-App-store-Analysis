```python
#Category or Genre plot
plt.figure(figsize=(15,15))
label=play_store.Category.value_counts().index
size=play_store.Category.value_counts().tolist()
plt.pie(size,labels=label)
plt.show()


#1746 - FAMILY - 18.13%
#1097 - GAME - 11.39%
#733 - TOOLS - 7.6 
```

```python
# Ratings plat
plt.figure(figsize=(20,20))
ratings = play_store.Rating.value_counts().index
size=play_store.Rating.value_counts().tolist()
plt.pie(size,labels=ratings)
plt.show()
```

```python
#Ratings
tot={'ratings':play_store.Rating.value_counts().index,
    'No of rated apps':play_store.Rating.value_counts().tolist()}
tot=pd.DataFrame(tot)
ratings = play_store.Rating.value_counts().index
size=play_store.Rating.value_counts().tolist()
tot.plot('ratings','No of rated apps',kind='bar')
plt.show()
```

```python
#Downloads accourding to price
plt.figure(figsize=(100,100))
dic={'price':play_store.Price.value_counts().index,
    'value':play_store.Price.value_counts().tolist()}
df=pd.DataFrame(dic)
df.plot('price','value',kind='line')
plt.show()
```

```python
#Downloads of kenrel density
plt.figure(figsize=(100,100))
dic={'price':play_store.Price.value_counts().index,
    'value':play_store.Price.value_counts().tolist()}
df=pd.DataFrame(dic)
df.plot('price','value',kind='kde')
plt.show()
```
